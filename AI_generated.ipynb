The following code asks the user to write any texts and gives the confidence score of how much of the text is AI Generated,the BART large MNLI model is used for zero-shot classification

```python
!pip install transformers
!pip install openai
from transformers import pipeline, set_seed

# Load the zero-shot classification pipeline with a specific model
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# Set the candidate labels
labels = ["Human", "AI"]

# Set the seed for reproducibility
set_seed(42)

# Get user input
text = input("Enter a piece of text: ")

# Classify the input text
result = classifier(text, labels)

# Print the classification result
if result['labels'][0] == 'Human':
    print(f"The input text is likely to be human-generated with a confidence score of {result['scores'][0]*100:.2f}%")
else:
    print(f"The input text is likely to be AI-generated with a confidence score of {result['scores'][1]*100:.2f}%")
```
